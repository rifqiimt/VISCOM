```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import streamlit
```

```python
# Baca gambar dan konversi ke grayscale
image = cv2.imread('tengkorak.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# ------------------ Metode 1: Histogram Equalization (HE) ------------------
he = cv2.equalizeHist(gray)

# ------------------ Metode 2: CLAHE ------------------
clahe_obj = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe = clahe_obj.apply(gray)

# ------------------ Visualisasi ------------------
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(gray, cmap='gray')
plt.title('Original Grayscale')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(he, cmap='gray')
plt.title('Histogram Equalization')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(clahe, cmap='gray')
plt.title('CLAHE')
plt.axis('off')

plt.tight_layout()
plt.show()
```
